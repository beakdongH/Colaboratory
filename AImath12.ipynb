```python
def diff(f,x):
  h = 1e-5
  return (f(x + h) - f(x - h)) / (2 * h)

#1변수 함수에 적용
def f(x):
  return x*(x-1)*(x-2)*(x-3)

#1변수함수 검사하강법
#학습률
r=0.01
x=2
for i in range(100):
  x -= r*diff(f,x)
  print(x,(f(x)))
```

```python
import numpy as np
#2변수함수에 적용
def g(x,y):
  return np.sin(x)+np.sin(y)

#2변수함수 검사하강법

def diff3D(f,sp):
  result=[]
  #g를 sp에서 x로 편미분
  def gx(x):
    return f(x,sp[1])
  result.append(diff(gx,sp[0]))

  #g를 sp에서 y로 편미분
  def gy(y):
    return f(sp[0],y)
  result.append(diff(gy,sp[1]))
  return np.array(result)

#학습률
r=0.04
sp = np.array([0.37,1.28])
for i in range(1000):
  sp -= r*diff3D(g,sp)
  print(sp,g(sp[0],sp[1]))
```

```python
import sympy as sp

a,b = sp.symbols("a b")
#추세선 찾기


#y=a+b
def f(x,a,b):
  return a*x+b
#예측값 list
y_pred = f(xlist,a,b)

#실제값 - 예측값 = 오차
errlist = ylist - y_pred

#오차제곱 평균 MSE
MSE = sum(errlist**2) / len(errlist)


def h(x,y):
  return float(MSE.subs([(a,x),(b,y)]))

h(1,3)

#학습률
r = 0.01

#시작점

sp = np.array([4,35.])
for i in range(100):
  sp -= r*diff3D(h,sp)
  print(sp,h(sp[0],sp[1]))
```
