```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
import os

#sampling rate를 1초에 몇개로 쪼개느냐
Fs = 48000.0

tlen = 1
Ts = 1/Fs

t = np.arange(0,tlen,Ts)
print(len(t))
x_list= np.arange(0,tlen,Ts)
y1_list= np.sin(2*np.pi*262*x_list)
y2_list= np.sin(2*np.pi*330*x_list)
y3_list= np.sin(2*np.pi*392*x_list)

y_list = y1_list+y2_list+y3_list
plt.plot(x_list[:2000],y_list[:2000],color='blue')
```

```python
scaled = np.int16(y_list*32767)

write('330.wave',48000,scaled)
```
